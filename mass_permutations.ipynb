```python
import numpy as np
import matplotlib.pylab as plt
```

Generate two random distributions from 0.5-1.5, representing two mass permutations (++-) for background events. 

We assume flat distributions for the background events.

```python
npts = 1000000

mass0 = 0.5 + np.random.random(npts)
mass1 = 0.5 + np.random.random(npts)

plt.figure()
h0 = plt.hist(mass0,bins=50,histtype='stepfilled')
plt.xlim(0.5,1.5)
plt.xlabel('Mass0')

plt.figure()
h1 = plt.hist(mass1,bins=50,histtype='stepfilled')
plt.xlim(0.5,1.5)
plt.xlabel('Mass1')

plt.figure()
h2 = plt.hist2d(mass0,mass1,bins=50)
```

Now, for each pair of numbers, select the one closest to 1.115 and plot those values.

```python
lam_candidates = np.zeros(len(mass0))

mass0_minus_lam = np.abs(mass0-1.115)
mass1_minus_lam = np.abs(mass1-1.115)

lam_candidates[mass0_minus_lam<=mass1_minus_lam] = mass0[mass0_minus_lam<=mass1_minus_lam]
lam_candidates[mass1_minus_lam<=mass0_minus_lam] = mass1[mass1_minus_lam<=mass0_minus_lam]

figure()
h3 = plt.hist(lam_candidates,bins=50,histtype='stepfilled')
plt.xlabel('Masses closest to Lambda (1.115)')
```

Plot with a finer binning and zoom in on a region similar to those in Fig. 3.32 - 3.41 in the analysis note. While those figures in the AN are the missing mass off the $K^+$ (the ``other" positive track), it just allows us to compare the mass range over which we will be looking for a $\Lambda$ peak.

```python
figure()
h4 = plt.hist(lam_candidates,bins=500,histtype='stepfilled')
plt.xlabel('Masses closest to Lambda (1.115)')
plt.xlim(0.9,1.3)
```
